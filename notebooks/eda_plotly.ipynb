```python
# --- EDA con Plotly: IBM Stock 1980-2025 ---

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Cargar dataset
df = pd.read_csv('../data/IBM_Stock_1980_2025.csv')

# --- Informaci√≥n general ---
print("üîç Dimensiones:", df.shape)
print("\nüß© Tipos de datos:\n", df.dtypes)
print("\nüìà Valores nulos:\n", df.isnull().sum())

# Convertir fecha si aplica
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# --- Estad√≠sticas b√°sicas ---
display(df.describe())

# --- Evoluci√≥n temporal del precio ---
fig = px.line(df, x='Date', y='Close', title='Evoluci√≥n del Precio de Cierre IBM')
fig.update_traces(line_color='royalblue')
fig.show()

# --- Volumen negociado ---
fig = px.area(df, x='Date', y='Volume', title='Volumen Negociado Diario', color_discrete_sequence=['orange'])
fig.show()

# --- Relaci√≥n entre precios ---
fig = px.scatter(df, x='Open', y='Close', color='Volume',
                 title='Relaci√≥n entre Precio de Apertura y Cierre',
                 color_continuous_scale='Viridis')
fig.show()

# --- Boxplot de precios ---
for col in ['Open', 'High', 'Low', 'Close']:
    fig = px.box(df, y=col, title=f'Distribuci√≥n de {col}')
    fig.show()

# --- Correlaciones num√©ricas ---
corr = df.corr(numeric_only=True)
fig = px.imshow(corr, text_auto=True, color_continuous_scale='RdBu_r', title="Mapa de Correlaci√≥n de Variables Num√©ricas")
fig.show()

# --- An√°lisis de tendencia y volatilidad ---
df['Daily_Change'] = df['Close'].pct_change() * 100
fig = px.line(df, x='Date', y='Daily_Change', title='Variaci√≥n Diaria (%) del Precio de Cierre')
fig.show()

# --- Distribuci√≥n de rendimientos ---
fig = px.histogram(df, x='Daily_Change', nbins=100, title='Distribuci√≥n de Rendimientos Diarios (%)')
fig.show()

print("‚úÖ EDA completado con visualizaciones interactivas.")
```
